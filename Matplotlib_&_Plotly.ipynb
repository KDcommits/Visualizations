# Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly
import plotly.express as px
import plotly.graph_objs as go
import plotly.offline as py
import plotly.tools as tls
from plotly.offline import iplot
import numpy as np 
import cufflinks as cf
import warnings 
warnings.filterwarnings('ignore')
```

###                                         Line Plot

```python
x=range(1,11)
y1=[1,6,5,6,8,8,4,2,15,10]
y2=[-2,1,8,10,3,6,9,4,8,7]
y3=[5,7,11,3,13,10,16,12,10,14]


plt.plot(x,y1,label='First_line',color='g',linewidth=2.5)
plt.plot(x,y2,label='Second_line',color='k',linestyle='--',marker='s',markerfacecolor='b',markersize=7.5,alpha=0.6)
plt.plot(x,y3,label='Third_line',color='y',marker='o',markeredgecolor='r',markeredgewidth=0.75)


plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Plot')
plt.tight_layout()
#plt.axis([0,12,-5,20])                       # To set the x and y axis limits in one line 
plt.xlim([0,12])                              # To set the x axis limit
plt.ylim([-5,20])                             # To set the y axis limit


plt.grid(True)                               
plt.legend(loc='lower center',frameon=True,fancybox=True,framealpha=0.5,borderpad=1,shadow=True,ncol=3,)                  # To determine where the legend is to be 

plt.savefig('Lineplot.png',dpi=300,bbox_inches='tight')    #bbox_inches: The portion of the figure to save. If 'tight' is passed, will attempt to trim the empty space around the figure
```

### Bar Plot(Using Matplotlib)

```python
x=range(1,6)
y=[3,3.3,4,4.5,6]

plt.bar(x,y,color='#AE67FF',width=0.65,edgecolor='w',linewidth=1.5,align='edge',hatch='///')


plt.xlabel('Experience(in Years)')
plt.ylabel('Salary(in LPA)')
plt.title('Salary of a Python developer')
plt.tight_layout()

plt.savefig('Barplot.png')
```

### Grouped bar-ploy (matplotlib)

```python
x=np.arange(5)
y1=[100,200,300,500,800]
y2=[80, 120, 280,420,640]
x_labels=['monday','tuesday','wenedsday','thursday','friday']
width=0.40
plt.bar(x-0.20,y1, width)
plt.bar(x+0.20,y2, width)
plt.axes().set_xticks(x)
plt.axes().set_xticklabels(x_labels)
plt.title("Grouped bar-plot in matplotlib")
plt.tight_layout()
```

# Plotly 

### Bar Plot

```python
symptoms={'symptom':['Fever','Dry cough','Fatigue','Sputum production','Shortness of breath',
        'Muscle pain','Sore throat','Headache','Chills','Nausea or vomiting','Nasal congestion',
        'Diarrhoea','Haemoptysis','Conjunctival congestion'],
          
        'percentage':[87.9,67.7,38.1,33.4,18.6,14.8,13.9,13.6,11.4,5.0,4.8,3.7,0.9,0.8]}

DF=pd.DataFrame(data=symptoms,index=range(14))

cf.getThemes()
cf.set_config_file(theme='solar')

fig = px.bar(DF[['symptom', 'percentage']].sort_values('percentage', ascending=False), 
             y="percentage", x="symptom", color='symptom', 
             log_y=True,title='Symptom of  Coronavirus')

fig.show()
```

###                                             Grouped Barplot

```python
df1 = px.data.tips()
fig = px.bar(df1, x="sex", y="total_bill", color='smoker', barmode='group',
             height=400)
fig.show()
```
