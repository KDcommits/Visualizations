### Histogram

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objs as go
import plotly.offline as py

df=sns.load_dataset("Titanic")         #sns.get_dataset_names() to get names of all the available dataset
df.head()

# Shows the Probability Density fn(PDF)

sns.distplot(df['age'],bins=40,hist=True,kde=True,rug=False,color='teal',
             vertical=False,norm_hist=False,label='Density')

                     #customization 
    
plt.title('Histogram',fontsize=14)
plt.xlabel('Age of Passengers',fontsize=13)
plt.ylabel('% of total Density',fontsize=13)
plt.xlim([-5,100])
plt.grid(False)
plt.xticks(np.arange(0, 100, step=10))    # Setting x-tick frequency    
sns.set_style('whitegrid')                # darkgrid, whitegrid, dark, white, ticks
sns.despine(trim=False,left=True)         # To remove axis spines
plt.tight_layout()
#sns.axes_style()                         # To customize the axis
```

### Barplot

EDA with Univariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df=sns.load_dataset("iris")       
df.head()
df_setosa=df.loc[df['species']=='setosa']
df_virginica=df.loc[df['species']=='virginica']
df_versicolor=df.loc[df['species']=='versicolor']

plt.plot(df_setosa['petal_length'],np.zeros_like(df_setosa['petal_length']),'o',label='Setosa')
plt.plot(df_virginica['petal_length'],np.zeros_like(df_virginica['petal_length']),'o',label='Virginica')
plt.plot(df_versicolor['petal_length'],np.zeros_like(df_versicolor['petal_length']),'o',label='Versicolor')

plt.xlabel('Petal_Length')
plt.xlim([0,8])
plt.title('EDA With Univariate analysis')
plt.legend(frameon=True)
plt.grid(False)
plt.tight_layout()
```

EDA With Bivariant Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df=sns.load_dataset("iris") 
df.head()

df_setosa=df.loc[df['species']=='setosa']
df_virginica=df.loc[df['species']=='virginica']
df_versicolor=df.loc[df['species']=='versicolor']

plt.scatter(df_setosa['petal_length'],df_setosa['petal_width'],color='red',label='Setosa')
plt.scatter(df_virginica['petal_length'],df_virginica['petal_width'],color='blue',label='Virginica')
plt.scatter(df_versicolor['petal_length'],df_versicolor['petal_width'],color='yellow',label='Versicolor')

plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('EDA with Bivariant Analysis')
plt.xlim([0,8])
plt.ylim([0,3])
plt.grid(False)
plt.legend(loc='upper left')
plt.tight_layout()
```
